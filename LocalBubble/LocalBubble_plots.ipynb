```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('thomas')
```

```python
import healpy as hp
```

```python
from astropy.io import fits
```

```python
i = 3

L19_0 = hp.read_map('L19_map-inner_final.fits',field=0)
vec_0 = np.asarray(hp.pix2vec(128,np.arange(hp.nside2npix(128))))
xyz_10_0 = vec_0 * L19_0

L19 = hp.read_map('L19_map-inner_final.fits',field=i)
vec = np.asarray(hp.pix2vec(128,np.arange(hp.nside2npix(128))))
xyz_10 = vec * L19

# and to roughly plot a cross-cut corresponding to the plane z = 0 (threshold to plot points 0.5 pc)
th = 0.5

plt.figure(figsize=(8,8))
plt.plot(xyz_10_0[0][np.abs(xyz_10_0[2])<th],xyz_10_0[1][np.abs(xyz_10_0[2])<th],'.');
plt.plot(xyz_10[0][np.abs(xyz_10[2])<th],xyz_10[1][np.abs(xyz_10[2])<th],'r.');

plt.plot(0,0,'y*',markersize=20)

plt.xlabel('x [pc]')
plt.ylabel('y [pc]')
plt.xlim(-300,300)
plt.ylim(-300,300)
plt.grid(linestyle='--')
```

```python
plt.figure(figsize=(8,8))

levels = np.linspace(-300,300,100)

colours = np.interp(levels, (levels.min(), levels.max()), (0, 1))

for i,j in zip(levels,colours):
    plt.plot(xyz_10[0][np.abs(xyz_10[2]-i)<th],xyz_10[1][np.abs(xyz_10[2]-i)<th],'.',alpha=0.5,color=plt.cm.magma(j));

plt.xlabel('x [pc]')
plt.ylabel('y [pc]')
plt.xlim(-300,300)
plt.ylim(-300,300)
plt.grid(linestyle='--')
```
